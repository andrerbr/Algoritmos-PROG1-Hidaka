```python
# Exemplo 01
# 1
lista = ['professor','renato','hidaka','torres']
def mapeamento(item):
  if item[0] == "R" or item[0] == 'r' or item[0] == 't' or item[0] == 'T':
    return item.upper()
  return item.lower()
lista = [mapeamento(item) for item in lista]
print(lista)
```

```python
#2
lista = ['professor','renato','hidaka','torres']
def mapeamento(item):
  if item[0] in {'r','R','t','T'}:
    return item.upper()
  return item.lower()
lista = [mapeamento(item) for item in lista]
print(lista)
```

```python
# Exemplo 02
lista = [1, 4, 13, 12, 8, 4, 5, 7, 9]

for indice, item in enumerate(lista):
    if item % 2 == 0:
        lista[indice] = item * 2
    else:
        lista[indice] = item / 2

print(lista)
```

```python
# 1
lista = ['Renato', 7, 17, 'Hidaka', 12, 9]

def filtro1(item):
    if type(item) is not str:
        return item
filtrados = [item for item in lista if filtro1(item)]
print(filtrados)
```

```python
# 1
cadastros = [
    {
      'nome': 'Renato Hidaka Torres',
      'e-mail': 'renatohidaka@gmail.com'

    },
    {
        'nome': 'Júlia Torres',
        'e-mail': 'julia@gmail.com'
    },
    {
      'nome': 'Maria José Silva',
      'e-mail': 'maria.silva@yahoo.com.br'

    },
    {
        'nome': 'Pedo Carlos Simões',
        'e-mail': 'pedrosimoes@gmail.com'
    }
]
def filtro02(item):
  if item['e-mail'].endswith('@gmail.com'):
    return item
filtrados = [item for item in cadastros if filtro02(item)]
print(*filtrados, sep = '\n')
```

```python
# List Comprehensions FILTRO e MAPEAMENTO
# EXEMPLO 01
cadastros = [
    {
      'nome': 'Renato Hidaka Torres',
      'e-mail': 'renatohidaka@gmail.com'

    },
    {
        'nome': 'Júlia Torres',
        'e-mail': 'julia@gmail.com'
    },
    {
      'nome': 'Maria José Silva',
      'e-mail': 'maria.silva@yahoo.com.br'

    },
    {
        'nome': 'Pedo Carlos Simões',
        'e-mail': 'pedrosimoes@gmail.com'
    }
]
def filtro01(item):
  return '@gmail' in item['e-mail']
def mapeamento(item):
  return item['nome']
filtrados = [mapeamento(item) for item in cadastros if filtro02(item)]
print(*filtrados, sep = '\n')
```

```python
# EXEMPLO 02
lista  = [10, 9, 11, 14, 8, 4, 7, 16, 12, 14]
nova_lista = []

for pred, item in zip(lista, lista[1:]):
    if item > pred:
        if item%2 == 0:
            nova_lista.append(item+1)
        else:
            nova_lista.append(item-1)

print(nova_lista)
```
