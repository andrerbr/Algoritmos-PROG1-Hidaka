```python
def contar_positivo (lista):
  n = 0
  for c in lista:
    if c > 0:
     n = n + 1
  return n
```

```python
def contar_menormedia (lista):
  n = 0
  soma = 0
  for c in lista:
    soma = soma + c
  media = soma/c
  for c in lista:
    if c < media:
      n = n + 1
  return media,n
```

```python
def contar_impares(lista):
  n = 0
  for c in lista:
    if (c % 2) == 1:
      n = n + 1
  return n
  
```
