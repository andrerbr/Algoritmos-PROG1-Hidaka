```python
def string_caixa_alta(lista):
  try:
    for indice, item in enumerate(lista):
      lista[indice] = item.upper()
    return lista
  except Exception:
    if item is not str:
      print('Digite apenas Strings')
```

```python
def string_tamanho(lista):
  try:
    for indice, item in enumerate(lista):
      lista[indice] = len(item)
    return lista
  except Exception:
    if item is not str:
      print('Digite apenas Strings')
```

```python
lista = [4,'andre','gomes','ribeiro']
a = string_tamanho(lista)
print(a)
```
