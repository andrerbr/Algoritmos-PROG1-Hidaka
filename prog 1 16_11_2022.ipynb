```python
# Função map
# se preocupar apenas a função de transformação do mapeamento
```

```python
def f(x):
  return x*2
```

```python
lista = [1,2,3,4,5]
lista = [f(item) for item in lista]
print(lista)
```

```python
# refatoração
lista = [1,2,3,4,5]
lista = list(map(f,lista))
print(lista)
```

```python
# criação de dicionario usando map
def funcao(chave,valor):
    return (chave,valor) 
produtos = ['Banana', 'Morango', 'Açaí', 'Laranja']
quantidade = [10, 13, 5, 12]
produtos = dict(map(funcao, produtos, quantidade))
print(produtos)
```

```python
# iterar as litas e que os os elementos sejam a soma das colunas
def soma(a,b,c):
  return a+b+c
lista_a = [3,7,6,5,4]
lista_b = [3,8,1,4,2]
lista_c = [2,7,2,9,9]

lista = list(map(soma, lista_a, lista_b, lista_c))
print(lista)
```

```python
# iterar as litas e que os os elementos sejam a soma das colunas com apenas um parametro
def soma(*arg):
  return sum(arg)
lista_a = [3,7,6,5,4]
lista_b = [3,8,1,4,2]
lista_c = [2,7,2,9,9]

lista = list(map(soma, lista_a, lista_b, lista_c))
print(lista)
```

```python
lista = ['6','8','9','99']
lista = list(map(int,lista))
print(lista)
```

```python
# Função anonima
def f(x):
  return x*2
lista = [1,2,3,4,5]
lista = list(map(lambda x: x*2, lista))
print(lista)
```

```python
produtos = ['Banana', 'Morango', 'Açaí', 'Laranja']
quantidade = [10, 13, 5, 12]
produtos = dict(map(lambda chave, valor:(chave,valor), produtos, quantidade))
print(produtos)
```

```python
#anonima
lista_a = [3,7,6,5,4]
lista_b = [3,8,1,4,2]
lista_c = [2,7,2,9,9]
lista_d = [2,7,2,9,9]
lista = list(map(lambda *arg: sum(arg), lista_a,lista_b,lista_c,lista_d))
print(lista)
```

```python
# list_comprehensions
lista_a = [3,7,6,5,4]
lista_b = [3,8,1,4,2]
lista_c = [2,7,2,9,9]
lista_d = [2,7,2,9,9]
lista = [sum(itens) for itens in zip(lista_a,lista_b,lista_c,lista_d)]
print(lista)
```

```python
# funcao filter
# funcao retornar funcao logico(true ou false)
def filtro(item):
  '''
  so positivos
  '''
  return item>=0
lista = [19, 6, 5, 6, -1, -5, 7, -8]
lista = list(filter(filtro,lista))
print(lista)
```

```python
filtro = lambda item: item[0].lower() == item[-1].lower()
lista = ['arara','cachorro','LoL','a','10302302302','9101010201219']
lista = list(filter(filtro,lista))
print(lista)
  
```
