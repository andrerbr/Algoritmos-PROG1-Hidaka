```python
def somar(x,y):
  resultado = x + y
  return resultado
def subtrair(x,y):
  resultado = x - y
  return resultado
def multiplicar(x,y):
  resultado = x * y
  return resultado
```

```python
# Funções
def modulo_1():
  valor1 = int(input("Digite o primeiro número: "))
  valor2 = int(input("Digite o segundo número: "))
  soma = somar(valor1,valor2)
  print(valor1, "+", valor2, "=",soma)
def modulo_2():
  valor1 = int(input("Digite o primeiro número: "))
  valor2 = int(input("Digite o segundo número: "))
  subtracao = subtrair(valor1,valor2)
  print(valor1, "-", valor2, "=",subtracao)
def modulo_3():
  valor1 = int(input("Digite o primeiro número: "))
  valor2 = int(input("Digite o segundo número: "))
  multi = multiplicar(valor1,valor2)
  print(valor1, "x", valor2, "=",multi)
def menu():
  op = input("1- Somar \n2- Subtrair\n3- Multiplicar\n0- Sair\n")
  return op
```

```python
# Programa Principal
def programa():
  controle = menu()
  while controle != "0":
    if controle == "1":
      modulo_1()
    elif controle == "2":
      modulo_2()
    elif controle == "3":
      modulo_3()
    else:
      print("Opção Inválida!")
    print("")
    controle = menu()
```

```python
programa()
```
