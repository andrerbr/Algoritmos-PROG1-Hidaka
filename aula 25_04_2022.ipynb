```python
# QUESTÃƒO 14 DA LISTA
def qt_14(lnomes,lidades,i,f):
  L_Saida = []
  N = len(lnomes)
  for k in range(N):
    if  lidades[k]>= i and lidades[k] <= f:
      L_Saida.append(lnomes[k])
  return L_Saida
```

```python
qt_14(['Ramon', 'Arnaldo', 'Raquel', 'Pedro', 'Rafael'],[23, 45, 27, 60, 45],20,30)
```

```python
L = [2,4,5,6,1,10]
soma = 0
N = len(L) # tamanho da lista / funcao
for k in range(N): # range = uma lista de 0 ate n-1 // posso iterar com k || n-1 = iterar ate o penultimo indice  || range(1) = um elementos // indice 0
    soma = soma + L[k]
print(soma)
```
