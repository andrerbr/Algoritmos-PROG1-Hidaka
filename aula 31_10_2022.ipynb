```python
#gerar lista
import random as r
def gerar_lista(tamanho):
  lista = []
  for i in range(tamanho):
    lista.append(r.randint(1,100))
  return lista
```

```python
# 1 - crie uma lista com 100 elementos tal que cada elemento seja um numero sorteado entre 1 e 100. informe os elementos distintos da lista
lista = gerar_lista(100)
conjunto = set(lista)
print(len(lista),len(conjunto))
```

```python
# 2 - crie duas listas com 100 elementos tal que cada elementos seja um numero sorteado entre 1 e 100. informe:
lista1 = gerar_lista(100)
lista2 = gerar_lista(100)
conjunto1 = set(lista1)
conjunto2 = set(lista2)
# a) os elementos que tem na lista 2, mas não tem na lista 1
# b) os elementos que não tem na lista 2, mas tem na lista 1
# c) os elementos que tem na lista 2 e na lista 1
# d) os elementos que tem na lista 1 e na lista 2, mas não em ambos
# e) os elementos que tem na lista 1 ou na lista 2
# f) quantos elementos eu tenho que remover de cada lista para que os conjuntos da lista sejam disjuntos
```

```python
# 2 LETRA A
print(conjunto2.difference(conjunto1))
```

```python
# 2 LETRA B
print(conjunto1.difference(conjunto2))
```

```python
# 2 LETRA C
print(conjunto1 & conjunto2)
```

```python
# 2 LETRA D
print(conjunto1 ^ conjunto2)
```

```python
# 2 LETRA E
print(conjunto1 | conjunto2)
```
