```python
# escreva um programa para receber uma lista l e um valor x e informar quantos elementos da lista l sÃ£o maiores ou igual a x
```

```python
def qtd_maior (l,x):
  qtd = 0
  for k in l:
    if k >= x:
      qtd = qtd + 1
  return qtd
```

```python
qtd_maior([1,4,5,6,7,8,9,10], 7)
```
