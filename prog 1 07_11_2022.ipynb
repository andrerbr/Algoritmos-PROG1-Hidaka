```python
# 1 - não permitir telefones iguais
# 2 - remover telefone
# 3 - função atualizar 
cadastros = []
def get_indice(telefone):
  for idx,item in enumerate(cadastros):
    if item['telefone'] == telefone:
      return idx
    return -1

def cadastrar():
    """Função para cadastrar o nome e telefone do usuário"""
    nome = input('Digite o nome:\n')
    telefone = input('Digite o telefone:\n')
    if get_indice(telefone) == -1:
      cadastros.append({'nome': nome, 'telefone': telefone})
      print('Usuário cadastrado com sucesso!\n')
    else:
      print('Telefone já existe')

def remover():
  telefone = input('Digite o telefone:\n')
  idx = get_indice(telefone)
  if idx == -1:
    cadastros.pop(idx)
    print('Usuario removido com sucesso \n')
  else:
    print('telefone ja existe!')
def listar_cadastros():
    """Função para listar o nome e telefone dos usuários casdastrados"""
    for item in cadastros:
        print('Nome:', item['nome'])
        print('Telefone:', item['telefone'])
        print()
    
def menu():
    """Função para ler a opção de menu escolhida pelo usuário"""
    print('Escolha um opção')
    print('1- listar cadastros')
    print('2- cadastrar')
    print('3- sair')
    return input()

def programa():
    """Função principal do programa"""
    opcao = menu()
    while opcao != '3':
        
        if opcao == '1':
            listar_cadastros()
        elif opcao == '2':
            cadastrar()
        else:
            print('Opção inválida!')
        
        opcao = menu()
```
