```python
def fatorial(N):
  fat = 1
  while N > 1:
    N, fat = N-1, fat*N
  return fat
```

```python
# refatoração com recursividade(eliminar iteração), saber quando para(if)
def fatorial_v2(n, fat = 1):
  if n == 1:
    return fat
  return fatorial_v2(n-1, fat*n)
```

```python
print(fatorial(3),fatorial_v2(3),fatorial_v3(5))
```

```python
def fatorial_v3(n):
  if n == 1:
    return n
  return n*fatorial_v3(n-1)
```

```python
# exemplo 2
def somar(n):
  soma = 0
  for x in range(1,n+1):
    soma = soma + 1
  return soma
```

```python
def somar_v2(n,soma = 0, x=1):
  if x == n+1:
    return soma
  return somar_v2(n,soma + x , x + 1)
```

```python
#exemplo 3
import random as r
def sorteio(n):
  repetido = True
  while repetido:
    lista = [r.randrange(1,101) for i in range(n)]
    repetido = len(lista) != len(set(lista))
  return lista
```

```python
import random as r
def sorteio_v2(n, repetido = True, lista = []):
  if not repetido:
    return lista
  lista = [r.randrange(1,101) for i in range(n)]
  repetido = len(lista) != len(set(lista))
  return sorteio_v2(n, repetido, lista)
```

```python
import random as r
def sorteio_v3(n):
  lista = [r.randrange(1,101) for i in range(n)]
  if len(lista) == len(set(lista)):
    return lista
  return sorteio_v3(n)
```

```python
# exemplo 4
def merge(e,d):
  v = []
  while len(e) > 0 and len(d) > 0:
    if e[0] <= d[0]:
      v.append(e[0])
      e.pop(0)
    else:
      v.append(d[0])
      d.pop(0)
    v.extend(e)
    v.extend(d)
    return v
```

```python
def merge_v2(e,d, v = []):
  if len(e) == 0 or len(d) == 0:
    v.extend(e)
    v.extend(d)
    return v
  if e[0] <= d[0]:
    v.append(e[0])
    e.pop(0)
    return merge_v2(e,d,v)
  else:
    v.append(d[0])
    d.pop(0)
    return merge_v2(e,d,v)
```
