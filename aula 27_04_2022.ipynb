```python
def qt_12(lnomes,lsaldo):
  l_saida = []
  N = len(lnomes)
  for k in range(N):
    if lsaldo[k] >= 0:
      l_saida.append(lnomes[k])
  return l_saida
```

```python
def qt_13(lnumeros,lsletras,r):
  l_saida = []
  N = len(lsletras)
  for k in range(N):
    if lnumeros[k] >= r:
      l_saida.append(lsletras[k])
  return l_saida
```
