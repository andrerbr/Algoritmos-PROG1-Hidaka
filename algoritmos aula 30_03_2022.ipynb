2) escreva um algoritmos para receber uma lista e informar quantos elementos são menores que a media.

3) escreva um algoritmo para receber uma lista e informar quantos elementos não são pares.

4) escreva um algoritmo para receber uma lista e informar quantos elementos são menores ou iguais ao tamanho da lista.

5) escreva um algoritmo para receber uma lista e informar quantos elementos são negativos.

```python
# 1 receber e armazenar a lista(ls)
# 2 soma = 0
# 3 N = 0 (quantidade de elementos)
# 4 iterar a lista (ls) com x
#  4.1 soma = soma + x
#  4.2 conta quantos elementos tem na lista (n = n +1)
# 5 dividir a soma dos elementos pela quantidade (media = soma/n)
# 6 exibir media

def calcular_media (ls):
  soma = 0
  n = 0
  for x in ls:
    soma = soma + x
    n = n + 1
  media = soma/n
  return media
```

```python
calcular_media([2,5, 3, 4])
```
