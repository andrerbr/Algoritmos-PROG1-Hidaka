```python
# Questão 50
def qt50(ls):
  soma = 0
  for x in ls:
    soma = soma + x
  return soma
def programa50():
  nome = input('Digite seu nome: ')
  saldo = int(input('Digite seu saldo: '))
  c1,c2,c3,c4,c5 = map(float, input('Digite 5 as compras: ').split())
  total_compras = qt50([c1,c2,c3,c4,c5])
  saldo_final = saldo - total_compras
  if saldo_final >= 0:
    print(nome,', seu saldo está positivo!')
    print('saldo atual',str('%.2f'%saldo_final))
  else:
    print(nome,', seu saldo está negativo!')
    print('saldo atual',str('%.2f'%saldo_final))
```

```python
# Questão 52
def programa52():
  valor_produto = float(input('Digite o valor do produto: '))
  desconto_produto = int(input('Digite o desconto do produto: '))
  saldo_bancario = float(input('digite o saldo bancario: '))
  desc = valor_produto*(desconto_produto/100)
  valor_final = valor_produto - desc
  if saldo_bancario > valor_final:
    print('compra pode ser realiza')
  else:
    print('não pode ser realizada')
```

```python
# Questão 62
def qt62(a,b):
  if a > b:
    maior = a
    menor = b
  else:
    maior = b
    menor = a
  return maior,menor
def programa62():
  a,b = map(int,input('Digite os valores de a e b: ').split())
  maior,menor = qt62(a,b)
  if maior % menor == 0:
    print('são multiplos')
  else:
    print('não são multiplos')
```

```python
# questao 66 incompleta
def qt66(ls):
  razao = ls[0] - ls[1]
  N = len(ls)
  for k in range(1,N):
    primeiro = ls[k-1]
    if  primeiro < ls[k] and (ls[k] - ls[k+1]) == razao:
      seq = print('Sequencia ordenada de forma crescente')
    elif  primeiro > ls[k] and (ls[k] - ls[k+1]) == razao:
      seq = print('Sequencia ordenada de forma decrescente')
    elif primeiro < ls[k] and (primeiro - ls[k]) != razao:
      seq = print('sequencia nao esta ordenada')
    elif  primeiro > ls[k] and (primeiro - ls[k]) != razao:
      seq = print('não esta ordenada')
    return seq
def programa66():
  a,b,c,d,e = map(int,input('Digite os Valores:').split())
  sequencia = qt66([a,b,c,d,e])
  print(sequencia)
```
