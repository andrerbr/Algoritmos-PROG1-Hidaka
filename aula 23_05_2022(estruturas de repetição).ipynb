```python
# Estruturas de repetição
# determinada => PARA => for
# inderteminada => enquanto => while
#   utiliza uma variavel sentinela
#     uma sentinela deve ser:   
#         1) inicializada antes da estrutura de repetição
#         2) ser utilizada na condição de parada da estrutura de repetição
#         3) ser atualizada no escopo da estrutura de repetição, tal que garanta a finitude da solução
```

```python
# Exemplo
def soma_2(x,y):
  soma = x + y
  return soma
```

```python
def programa():
  while True:
    x = int(input('Digite o primeiro número: '))
    y = int(input('Digite o segundo número: '))
    resultado = soma_2(x,y)
    print(x, '+', y, '=', resultado)
    c = int(input('1-sair, 0- continuar\n'))
    if c == 1:
      break
```

```python
programa()
```
