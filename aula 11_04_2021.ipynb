```python
# 1) receber e armazenar a lista em l
# 2) p = 0
# 3) i = 0
# 4) iterar a lista l com x
#  4.1) x mod 2 == 0? 
#  4.1.1) p = p + 1
# 5) exibir p,i
    
```

```python
  def impar_par(ls):
   p = 0
   i = 0
   for x in ls:
    if x % 2 == 0:
      p = p + 1
    else:
     i = i + 1
   return p,i 
```

```python
par,impar = impar_par([1,5,6,7,8,8,8,3,9])
print('a quantidade de numeros pares é:',par)
print('a quantidade de numeros impares é:',impar)
```
