```python
def questao02(lista,string = 'as'):
    try:
      l_string = []
      if not isinstance(lista, list) or not isinstance(string,str):
        raise None
      else:
        for x in lista:
          if x.endswith(string):
            l_string.append(x)
        return sorted(l_string, reverse = True)
    except Exception:
      return None
```

```python
lista = 1
q = questao02(lista,'o')
print(q)
```

```python
def qt03(lista):
        try:
          for indice, item in enumerate(lista):
            if not isinstance(item,str):
              raise None
            else:
              lista[indice] = item.capitalize()
          return lista
        except Exception:
          return None
            
```

```python
lista = ['ab','cd','de']
q = qt03(lista)
print(q)
```
