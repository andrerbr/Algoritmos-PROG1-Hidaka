```python
def qt1(ls):
  soma = 0
  lsaida = []
  N = len(ls)
  for k in range(1,N):
    maior = ls[k-1]
    if  ls[k] > maior:
      soma = soma + ls[k]
  lsaida.append(soma)
  return lsaida
```

```python
qt1([5,5,5,5,5,5,5,5])
```

```python
# 2
def qt2(ls,r):
  lsaida = []
  for x in ls:
    if x*2 > r:
      lsaida.append(x)
  return lsaida
```

```python
qt2([2,4,1,12,8,6,9],15)
```

```python
def qt3(ls,a,c):
  qtd = 0
  for x in ls:
    if x[0] == a or x[3] == c:
      qtd = qtd +1
  return qtd
```

```python
qt3(["TACG","GAGC","ATUC","TAGC","GAGC"],"A","C")
```

```python
# 4
def qt_4 (ls,i,f):
  qtd = 0
  for x in ls:
    if x < i or x > f:
      qtd = qtd + 1
  return qtd
```

```python
qt_4([2,33,21,24,25,6,7,80,73,55,60,5,40,29],20,35)
```
