```python
#Questão 2
import math as m
# codificação
def qtd_lajotas(ls,cs,ll,cl):
  ls = ls * 100
  cs = cs * 100
  qtdL = ls/ll
  qtdC = cs/cl
  qtdL = m.ceil(qtdL)
  qtdC = m.ceil(qtdC)
  qtd = qtdL * qtdC
  return qtd
# implementação
def programa_02():
  print('Programa 02: Calcula a quantidade de lajotas que Mário vai usar')
  ls, cs = (map(int,input('largura da sala e comprimento da sala:\n ').split()))
  ll, cl = (map(int,input('largura da lajota e comprimento da lajota:\n ').split()))
  quantidade = qtd_lajotas(ls,cs,ll,cl)
  print(quantidade)
```

```python
#Questão 21
# M = Px(i+j)ˆn
# P = valor inicial
# j = taxa de juros
# N = tempo
# M -> 2P = Px(i + j)ˆn
#      2P/P = (i+j)ˆn
#       2 = (i + j)ˆn
#       log2 = log(i+j)ˆn
#       log2 = nlog(i + j)
#       N = log2/log(i+j)

import math as m
def calcular_tempo(juros):
  juros = juros/100
  N  = m.log10(2) / m.log10(1+juros)
  return N

def programa_Q21():
  juros = float(input('informa a taxa de juros \n'))
  tempo = calcular_tempo(juros)
  print('é necessario', m.floor(tempo), 'meses')
```
