```python
def q05(ls):
  soma = 0
  for x in ls:
    soma = soma + x
  return soma
def programa05():
  x = [int(y) for y in input("digite o valor das contas").split()]
  r = int(input('digite o palpite: '))
  contat = q05(x)
  if r == contat:
    print('Acertou')
  else:
    print("errou")
```

```python
def q04(nota):
  if nota == 0:
    resultado = "E"
  elif nota >= 1 and nota <= 35:
    resultado = "D"
  elif nota >= 36 and nota <= 60:
    resultado = "C"
  elif nota >= 61 and nota <= 85:
    resultado = "B"
  elif nota >= 86 and  nota <= 100:
    resultado = "A"
  return resultado

def programa04():
  x = int(input("Digite a nota do aluno: "))
  result = q04(x)
  print(result)
```

```python
def programa02():
  n_band = int(input('Bandejas entregues:'))
  qtd = 0
  ncb = 0
  while True:
    nl,nc = map(int,input("Digite o numero de latas e copos ").split())
    if nl > nc:
      ncb = ncb + nc
    qtd = qtd + 1
    if qtd == n_band:
      break
  print(ncb)
```
