```python
# "Lista dentro de outra lista"

#1. receber e armazenar a lista em L
#2. S = [] 
#3. iterar a lista L com X
  #3.1 x inicia com a letra "A"
      #3.1.1 Adicionar x em S
#4. exibir S

##** append(x) =  adicionar elementos na lista vazia
```

```python
def listaA(L):
  S = []
  for x in L:
    if x[0] == 'A' or x[0] == 'a':
      S.append(x)
  return S
```

```python
listaA(["abacaxi","Uva","amado","arara","Jaca","Arca"])
```

```python
def qt_11(r,ls):
  qtd = []
  for x in ls:
    if x >= r:
      qtd.append(x)
  return qtd
```

```python
Pq = qt_11(8,[5, 4, 3, 11, 1, 1, 8 ,10, 11, 3, 4, 2, 10])
print("Produtos na validade = ", Pq)
```
