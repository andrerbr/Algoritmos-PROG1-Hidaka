```python
# Questão 90
def encontrar_id_maior(x):
  maior = x[0]
  indice = 0
  n = len(x)
  for y in range(n):
    if x[y] > maior:
        maior = x[y]
        indice = y
  return indice

def programa():
  x = [int(y) for y in input('Digite os valores: \n').split()]
  indice = encontrar_id_maior(x)
  print('O maior elemento está no índice', indice)
```

```python
programa()
```

```python
# Questão 86
def qt86(x):
  soma = 0
  n = len(x)
  for y in range(n):
    soma = soma + x[y]
  media = soma/n
  return media

def programa86():
  x = [int(y) for y in input('Digite os valores: \n').split()]
  media = qt86(x)
  media = str('%.2f'%media)
  print('MÉDIA = ',media)
```

```python
# QUESTÃO 91
def qt_91(ls):
  menor = ls[0]
  maior = ls[0]
  for y in ls:
    if y > maior:
      maior = y
    if y <= menor:
      menor = y
  return maior,menor
def menu():
  print('ESCOLHA UMA OPERAÇÃO!')
  op = input("1- Somar \n2- Subtrair\n3- Multiplicar\n4- Dividir\n0- Sair\n")
  return op
```

```python
def programa91():
  x = [int(y) for y in input('Digite os valores: \n').split()] 
  a,b = qt_91(x) 
  controle = menu()
  while controle != "0":
      if controle == "1":
          print('O Resultado de ',a,'+',b,'=',a+b)
          print('\n')
          controle = menu()
      elif controle == "2":
          print('O Resultado de ',a,'-',b,'=',a-b)
          print('\n')
          controle = menu()
      elif controle == "3":
          print('O Resultado de ',a,'x',b,'=',a*b)
          print('\n')
          controle = menu()
      elif controle == "4":
          print('O Resultado de ',a,'/',b,'=',a/b)
          print('\n')
          controle = menu()
      else:
          print("Opção Inválida!")
          print('\n')
          controle = menu()
```
