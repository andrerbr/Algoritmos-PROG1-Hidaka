```python
# menor valor da lista
# 1 - receber e armazenar a lista  em (l)
# 2 - menor = L[0]
# 3 - iterar a lista (l) com (x)
# 3.1 - x < menor?
# 3.1.1 - menor = x
# 4 - exibir menor
```

```python
def menor_lista(l):
  menor = l[0]
  for x in l:
    if x<menor:
      menor = x
  return menor
```

```python
menor_lista([5,9,8,10,3,8,96,9])
```

```python
# questÃ£o 3

def qt3 (al):
  mal = al[0]
  for x in al:
    if x > mal:
      mal = x
  return mal
```
