```python
#QUESTÃO 01
def q01(a,b):
  pg= a*312
  fpg= pg / b
  return fpg
```

```python
print("PREÇO DA GASOLINA")
x,y = map(float,input("Digite o preço da gasolina e a autonomia do carro: ").split())
pgf = q01(x,y)
pgf = str('%.2f'%pgf)
print(pgf)
```

```python
def q40(nota):
  ncem = int(nota / 100)
  ncemr = int(nota % 100)
  ncinq = int(ncemr/50)
  ncinqr = int(nota % 50)
  nvint = int(ncinqr/20)
  nvintr = int(nota % 10)
  ndez = int(nvintr/10)
  ncinc = int(nvintr/5)
  ncincr = int(nota%5)
  ndois = int(ncincr/2)
  num = int(ncincr%2)
  
  

  
  return ncem,ncinq,nvint,ndez,ncinc,ndois,num
```
