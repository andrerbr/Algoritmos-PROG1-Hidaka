```python
  # CONTINUACAO
# filtro com Funcao anonima
def filtro (item):
  for q in range(2,item):
    if item % q == 0:
      return False
  return True
lista = [3,15,17,14,101,98,45,13,7,16,21]
lista = list(filter(filtro,lista))
print(lista)
```

```python
# filter com anonima com all
filtro = lambda item: all(item % q != 0 for q in range(2,item))
lista = [3,15,17,14,101,98,45,13,7,16,21]
lista = list(filter(filtro,lista))
print(lista)
```

```python
# filter com anonima com any
filtro = lambda item: not any(item % q == 0 for q in range(2,item))
lista = [3,15,17,14,101,98,45,13,7,16,21]
lista = list(filter(filtro,lista))
print(lista)
```

```python
# funcao reduce
# manipulacao de colecao que o resultado desse manipulacao seja apenas um valor
lista = [3,15,17,14,101,98,45,13,7,16,21]
soma = 0
for item in lista:
  soma += item
print(soma)
```

```python
# exemplo reduce
from functools import reduce
lista = [3,15,17,10]
soma = reduce(lambda soma, valor: soma+valor,lista,0)
print(soma)
```

```python
# exemplo 01
from functools import reduce
lista = [30,10,6,7,9,10]
valor = reduce(lambda minimo, valor: valor if valor < minimo else minimo, lista)
print(valor)
```
