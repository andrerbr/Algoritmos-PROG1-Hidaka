```python
# QUESTÃO 65 LISTA 2
def caminho_b(p,r):
  if p == 1 and r == 1:
    caminho = 'A'
  elif p == 1 and r == 0:
    caminho = 'B'
  else:
    caminho = 'C'
  return caminho
```

```python
print("DESCOBRIR O CAMINHO DA BOLA")
a,b = map(int,input("Digite os valores de p e r: ").split())
f = caminho_b(a,b)
print(f)
```

```python
# OUTRA MANEIRA #2 DA QUESTÃO 37
def ehmaior2(a,b,c):
  if a>b and a>c:
    m = "Maior A"
  elif b>a and b>c:
    m = "Maior B"
  else:
    m = "Maior C"
  return m
```

```python
# OUTRA MANEIRA #3 DA QUESTÃO 37
def ehmaior_3(a,b,c):
  m1 = (a + b + abs(a-b)) / 2
  m2 = (m1 + c + abs(m1-c)) / 2
  return m2
```

```python
x,y,z = map(int,input().split())
f = ehmaior_3(x,y,z)
print(f) 
```
