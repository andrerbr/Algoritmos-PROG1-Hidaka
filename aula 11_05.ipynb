```python
# QUESTÃO 26
def soma(a, b):
  c = a + b
  return c
```

```python
# PSEUDOCODIGO
# LEIA(N1)
# LEIA(N2)
# R = |N1 + N2| ----> soma
# ESCREVA(R)
# -----------
# int() = inteiro
# floar() = real
print('***Programa para somar dois números!***')
n1 = float(input("Digite o primeiro número:\n"))
n2 = float(input("Digite o segundo número:\n"))
R = soma(n1,n2)
print("A soma é de",n1,"+",n2,"=",R)
```

```python
# QUESTÃO 27
def areac(r):
  a = 3.14159 * r**2
  return a
```

```python
print('programa da questão 27')
valor = float(input('Digite o valor do raio:'))
area = areac(valor)
area = str('%.2f'%area) # ----> limitar casas decimais 
print('A =', area)
```

```python
# QUESTÃO 37
def ehmaior(a,b,c):
  lista = [a,b,c]
  maior = lista[0]
  for x in lista:
    if x > maior:
      maior = x
  return maior
```

```python
print("Descobrir qual numero é maior")
n1,n2,n3 = map(int,input("Digite 3 numeros separados por espaço: ").split()) # mais de 1 variael em um mesmo input --> a,b,c = map(int,input().split())
s = ehmaior(n1,n2,n3)
print(s,'eh maior')
```
