```python
# base questão 2,3
lista = [4, 40, 50, 68, 28, 75, 73, 17, 55, 17]
nova_lista = []
for itens in zip(lista, lista[1:], lista[2:]):
  x = itens[0]
  for item in itens:
    if item < x:
      x = item
  nova_lista.append(x)
print(nova_lista)
```

```python
# QUESTÃO 2
lista = [4, 40, 50, 68, 28, 75, 73, 17, 55, 17]
lista = [min(itens) for itens in zip(lista, lista[1:],lista[2:])]
print(lista)
```

```python
# QUESTÃO 3
lista = [ 4,40,50,68,28,75,73,17,55,17]
nova_lista = list(map(min,zip(lista,lista[1:],lista[2:])))
print(nova_lista)
```

```python
# QUESTÃO 5
import re
lista = ['QWT997z', 'PoP', 'Raquete-X67', 'XYZw&', 'F99-bx', 'Hall']
nova_lista = list(filter(lambda item:re.fullmatch("[A-Z].*[a-z]",item),lista))
print(nova_lista)
```

```python
# Questão 9
from functools import reduce
lista = [
{
'produto': 'A',
'quantidade': 10
},
{
'produto': 'B',
'quantidade': 2
},
{
'produto': 'C',
'quantidade': 30
},
{
'produto': 'D',
'quantidade': 17
},
]
qtd = (reduce(lambda qtd, item: qtd + item['quantidade'],lista,0)/len(lista))
print(qtd)
```
